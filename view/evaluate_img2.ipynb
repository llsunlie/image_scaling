# Average enlarge / reduce

```python
from multi_evaluate import evaluate_v, evaluate_h
import numpy as np

ALGORITHM_NAME_LIST = [
    'bicubic',
    'bilinear',
    'cubicBSpline',
    'lagrange',
    'lanczos2',
    'lanczos3',
    'nearest',
    'newton',
    'shannon'
]
ALGORITHM_ARGS_MAP = {
    'lagrange': [3],
    'newton': [3],
    'shannon': [3]
}
SRC_PATH_LIST = [
    '../data/src/img2_90x160.png',
    '../data/src/img2_360x640.png',
    '../data/src/img2_720x1280.png',
    '../data/src/img2_1080x1920.png'
]
IMG_COUNT = len(SRC_PATH_LIST)

res = [[[] for _ in range(IMG_COUNT)] for _ in range(IMG_COUNT)]
for i in range(IMG_COUNT):
    for j in range(IMG_COUNT):
        if i == j:
            continue
        res[i][j] = evaluate_h(
            SRC_PATH = SRC_PATH_LIST[i],
            ORIGINAL_IMG_PATH = SRC_PATH_LIST[j],
            TARGET_DIR = '../data/target/',
            ALGORITHM_NAME_LIST = ALGORITHM_NAME_LIST,
            ALGORITHM_ARGS_MAP = ALGORITHM_ARGS_MAP,
            SHOW_LOG = False,
            SHOW_IMG = False,
            SHOW_BAR = False
        )
        print(f'res[{i}][{j}] = {res[i][j]}')
```

```python
from utils.chart_barh import ChartBarh, MultiChartBarH

enlarge_psnr_avg = np.zeros(len(ALGORITHM_NAME_LIST))
enlarge_ssim_avg = np.zeros(len(ALGORITHM_NAME_LIST))
enlarge_ms_ssim_avg = np.zeros(len(ALGORITHM_NAME_LIST))
enlarge_ws_psnr_avg = np.zeros(len(ALGORITHM_NAME_LIST))
enlarge_s_psnr_avg = np.zeros(len(ALGORITHM_NAME_LIST))
for i in range(IMG_COUNT):
    for j in range(i + 1, IMG_COUNT):
        psnr_list = [obj.psnr for obj in res[i][j]]
        ssim_list = [obj.ssim for obj in res[i][j]]
        ms_ssim_list = [obj.ms_ssim for obj in res[i][j]]
        ws_psnr_list = [obj.ws_psnr for obj in res[i][j]]
        s_psnr_list = [obj.s_psnr for obj in res[i][j]]
        enlarge_psnr_avg += psnr_list
        enlarge_ssim_avg += ssim_list
        enlarge_ms_ssim_avg += ms_ssim_list
        enlarge_ws_psnr_avg += ws_psnr_list
        enlarge_s_psnr_avg += s_psnr_list
enlarge_psnr_avg /= (IMG_COUNT * (IMG_COUNT - 1)) / 2
enlarge_ssim_avg /= (IMG_COUNT * (IMG_COUNT - 1)) / 2
enlarge_ms_ssim_avg /= (IMG_COUNT * (IMG_COUNT - 1)) / 2
enlarge_ws_psnr_avg /= (IMG_COUNT * (IMG_COUNT - 1)) / 2
enlarge_s_psnr_avg /= (IMG_COUNT * (IMG_COUNT - 1)) / 2

chart_barh_list = [
    ChartBarh(
        data=enlarge_psnr_avg,
        yticks=range(len(ALGORITHM_NAME_LIST)),
        yticklabels=ALGORITHM_NAME_LIST,
        xlabel='enlarge_psnr_avg',
        title='PSNR'
    ),
    ChartBarh(
        data=enlarge_ssim_avg,
        yticks=range(len(ALGORITHM_NAME_LIST)),
        yticklabels=ALGORITHM_NAME_LIST,
        xlabel='enlarge_ssim_avg',
        title='SSIM'
    ),
    ChartBarh(
        data=enlarge_ms_ssim_avg,
        yticks=range(len(ALGORITHM_NAME_LIST)),
        yticklabels=ALGORITHM_NAME_LIST,
        xlabel='enlarge_ms_ssim_avg',
        title='MS_SSIM'
    ),
    ChartBarh(
        data=enlarge_ws_psnr_avg,
        yticks=range(len(ALGORITHM_NAME_LIST)),
        yticklabels=ALGORITHM_NAME_LIST,
        xlabel='enlarge_ws_psnr_avg',
        title='WS_PSNR'
    ),
    ChartBarh(
        data=enlarge_s_psnr_avg,
        yticks=range(len(ALGORITHM_NAME_LIST)),
        yticklabels=ALGORITHM_NAME_LIST,
        xlabel='enlarge_s_psnr_avg',
        title='S_PSNR'
    )
]

MultiChartBarH(
    suptitle = 'ENLARGE',
    chart_barh_list = chart_barh_list,
    row = 2
).plot()
```
